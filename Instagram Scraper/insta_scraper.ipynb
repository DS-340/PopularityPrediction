```python
# 🚀 STEP 6: Move and rename results to keep accounts separate
import shutil
from datetime import datetime

# Define save path in Shared Drive
base_drive_path = "/content/drive/Shareddrives/DS340W Project"
timestamp = datetime.now().strftime("%Y-%m-%d_%H-%M")
account_dir = f"{base_drive_path}/{username}_{timestamp}"

os.makedirs(account_dir, exist_ok=True)

# Move and rename each results file
for file in os.listdir("results"):
    src = f"results/{file}"
    dst = f"{account_dir}/{username}_{file}"
    shutil.copy(src, dst)

print(f"Saved {username}'s results to: {account_dir}")
```

```python
# ✅ STEP 7: Read and preview the data nicely
import json
import pandas as pd

# Load from the saved path in Drive
json_path = f"{account_dir}/{username}_all-user-posts.json"

with open(json_path, "r") as f:
    data = json.load(f)

# Extract readable info
posts = []
for post in data:
    caption = post.get("caption", {}).get("text", "")
    likes = post.get("like_count", None)
    comments = post.get("comment_count", None)
    is_video = "Yes" if post.get("video_versions") else "No"

    posts.append({
        "Caption": caption,
        "Likes": likes,
        "Comments": comments,
        "Is Video": is_video
    })

# Display it
df = pd.DataFrame(posts)
df.head(10)
```

```python
# ✅ STEP 8 (Optional): Save table to CSV for future analysis
df.to_csv(f"{account_dir}/{username}_instagram_posts.csv", index=False)
print("📁 Saved as CSV:", f"{account_dir}/{username}_instagram_posts.csv")
```
